**Problema #2:**
Desarrolle una funci ́on que dada dos im ́agenes a color, realice las operaciones SUMA, RESTA, AND,
OR , XOR con ambas im ́agenes recuerde que para realizar las operaciones l ́ogicas es necesario pasar
la imagen a escala de grises y luego binarizarla

```python
imagen1 = cv2.imread('cora.png',0)
imagen2 = cv2.imread('circulo.png',0)

def problema2(imagen1, imagen2, operacion):
  #Transformacion de imagenes a escala de grises 
  ret,imagen1 = cv2.threshold(imagen1,127,255,cv2.THRESH_BINARY)
  imagen1 = cv2.cvtColor(imagen1, cv2.COLOR_BGR2RGB)
  ret,imagen2 = cv2.threshold(imagen2,127,255,cv2.THRESH_BINARY)
  imagen2 = cv2.cvtColor(imagen2, cv2.COLOR_BGR2RGB)

  if operacion == "suma":
    imagen = imagen1 + imagen2
    plt.imshow(imagen, 'gray')
    plt.show()
  if operacion == "resta":
    imagen = imagen1 - imagen2
    plt.imshow(imagen, 'gray')
    plt.show()
  if operacion == "and":
    imagen = imagen1 & imagen2
    plt.imshow(imagen, 'gray')
    plt.show()
  if operacion == "or":
    imagen = imagen1 |imagen2
    plt.imshow(imagen, 'gray')
    plt.show()
  if operacion == "xor":
    imagen = np.logical_xor(imagen1,imagen2)
    plt.imshow(imagen, 'gray')
    plt.show()
```

```python
problema2(imagen1, imagen2, "suma")
```

```python
problema2(imagen1, imagen2, "resta")
problema2(imagen2, imagen1, "resta")
```

```python
problema2(imagen1, imagen2, "and")
```

```python
problema2(imagen1, imagen2, "or")
```

```python
problema2(imagen1, imagen2, "xor")
```

**Problema #3:**
Desarrolle una funci ́on que dada una imagen en escala de grises muestre dicha imagen en 3 di-
mensiones, donde los ejes X e Y representan el dominio espacial de la imagen y Z representa la
intensidad del color gris de cada pixel.

```python
def problema3(img): 
    height = img.shape[0]
    width = img.shape[1]
    #x y y 
    x = np.linspace(0, width, width)
    y = np.linspace(0, height, height)   
    #creacion de mesh
    x, y = np.meshgrid(x,y)
    #impresion de figura
    fig = plt.figure(figsize=[12,10])
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, img)
    plt.show()
```

```python
prueba = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
problema3(prueba)
```
